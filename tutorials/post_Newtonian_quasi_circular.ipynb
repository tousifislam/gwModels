## Example notebook to use post-Newtonian calculations
This notebook uses calculations presented in https://arxiv.org/pdf/2304.11185 and references therein for quasi-cricular non-spinning binaries

```python
import warnings
warnings.filterwarnings("ignore", "Wswiglal-redir-stdio")
import matplotlib.pyplot as plt
import numpy as np

# import gwModels
import sys
PATH_TO_GWMODELS = "/Users/tousifislam/Documents/work/gwModels/" 
# PATH_TO_GWMODELS = "/home/tousifislam/Documents/works/git_repos/gwModels/"
sys.path.append(PATH_TO_GWMODELS)
import gwModels
```

```python
# mass ratio
q = 3

# access the model presented by Blanchet et al
# 4PN amplitude, 4.5 PN GW phase and 4.5 frequency evolution
obj = gwModels.pn.Blanchet2024(q=q)
```

```python
obj.plot_waveform()
```

```python
plt.plot(obj.t, obj.x['4.5'])
plt.ylim(-0.1,0.1)
plt.xlabel('Time', fontsize=18)
plt.ylabel('$x$', fontsize=22)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```

```python
plt.plot(obj.t, obj.phase_dict['4.5'])
plt.xlabel('Time', fontsize=18)
plt.ylabel('$\\phi$', fontsize=22)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=13)
plt.grid()
plt.show()
```
