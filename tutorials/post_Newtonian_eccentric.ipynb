## Example notebook to use post-Newtonian calculations for eccentric binaries

This notebook uses calculations presented in https://arxiv.org/pdf/1609.05933 and https://arxiv.org/pdf/0806.1037

```python
import warnings
warnings.filterwarnings("ignore", "Wswiglal-redir-stdio")
import matplotlib.pyplot as plt
import numpy as np

# import gwModels
import sys
PATH_TO_GWMODELS = "/Users/tousifislam/Documents/work/gwModels/" 
# PATH_TO_GWMODELS = "/home/tousifislam/Documents/works/git_repos/gwModels/"
sys.path.append(PATH_TO_GWMODELS)
import gwModels
```

### Generate PN waveforms

```python
q = 1
x0 = 0.07
e0 = 0.1
t0 = 0.0
phi0 = 0.0
l0 = 0.0
```

```python
# x-model waveform
wf_xmodel = gwModels.pn.ecc_nospin.xmodel(q, x0, e0, t0, l0, phi0)

# (lazy) EccentricIMR = x-model + NRSurrogate merger-ringdown
wf_hinder_imr = gwModels.pn.ecc_nospin.EccentricIMRHinder2017(q, x0, e0, t0, l0, phi0)
```

```python
# ax-model
wf_axmodel = gwModels.pn.ecc_nospin.axmodel(q, x0, e0, t0, l0, phi0)

# (lazy) ENIGMA = ax-model + NRSurrogate merger-ringdown
wf_imr = gwModels.pn.ecc_nospin.EccentricIMRWave(q, x0, e0, t0, l0, phi0)
```
